```python
DATASET_DIR = Path("dataset")
```

```python
df_item = pl.read_parquet(DATASET_DIR / "item_2024_clean.parquet")

df_item_clean2 = (
    df_item
    # drop item không có giá
    .filter(pl.col("price").is_not_null())
    # fill null cho categorical
    .with_columns([
        pl.col(c).fill_null("Unknown")
        for c in [
            "category_l1", "category_l2", "category_l3",
            "category", "brand", "gender_target", "age_group",
            "item_type", "color", "size", "origin", "material"
        ]
    ])
)

df_item_clean2.null_count()
```

```python
df_user = pl.read_parquet(DATASET_DIR / "user_2024_clean.parquet")

df_user_clean2 = (
    df_user
    .with_columns([
        pl.col("gender").fill_null("Unknown"),
        pl.col("province").fill_null("Unknown"),
        pl.col("region").fill_null("Unknown"),
        pl.col("membership").fill_null("Standard"),
        pl.col("install_app").fill_null("Unknown"),
    ])
)

df_user_clean2.null_count()
```

```python
df_purchase = pl.read_parquet(DATASET_DIR / "purchase_2024_clean.parquet")

df_purchase_clean2 = (
    df_purchase
    # drop interaction không hợp lệ
    .filter(
        pl.col("customer_id").is_not_null() &
        pl.col("item_id").is_not_null() &
        pl.col("price").is_not_null()
    )
    # fill số
    .with_columns([
        pl.col("quantity").fill_null(1),
        pl.col("discount").fill_null(0),
    ])
    # fill categorical
    .with_columns([
        pl.col(c).fill_null("Unknown")
        for c in ["channel", "payment"]
    ])
)

df_purchase_clean2.null_count()
```

```python
df_item_clean2.write_parquet(DATASET_DIR / "item_2024_ready.parquet")
df_user_clean2.write_parquet(DATASET_DIR / "user_2024_ready.parquet")
df_purchase_clean2.write_parquet(DATASET_DIR / "purchase_2024_ready.parquet")
```

xử lý outlier

```python
DATASET_DIR = Path("dataset")

df = pl.read_parquet(DATASET_DIR / "purchase_2024_user_ge2.parquet")
```

```python
df.select(["quantity", "price", "discount"]).describe()
```

```python
q99_qty = df.select(pl.col("quantity").quantile(0.99)).item()
p995_price = df.select(pl.col("price").quantile(0.995)).item()

q99_qty, p995_price
```

```python
df = df.with_columns(
    pl.col("quantity")
      .clip(1, q99_qty)
      .alias("quantity")
)
```

```python
df = df.with_columns(
    pl.when(pl.col("discount") < 0)
      .then(0)
      .when(pl.col("discount") > pl.col("price"))
      .then(pl.col("price"))
      .otherwise(pl.col("discount"))
      .alias("discount")
)
```

```python
df = df.filter(pl.col("price") > 0)

df = df.with_columns(
    pl.col("price")
      .clip(None, p995_price)
      .alias("price")
)
```

```python
df = df.with_columns(
    (
        pl.col("quantity").log1p()
        * (1 - pl.col("discount") / pl.col("price"))
    ).alias("implicit_score")
)
```

```python
s99 = df.select(pl.col("implicit_score").quantile(0.99)).item()

df = df.with_columns(
    pl.col("implicit_score")
      .clip(None, s99)
      .alias("implicit_score")
)
```

```python
df.select(["quantity", "price", "discount", "implicit_score"]).describe()
```

implicit_Score có 1 giá trị âm => xử lý

```python
df = df.with_columns(
    (
        pl.col("quantity").log1p()
        * (1 - pl.col("discount") / pl.col("price"))
    )
    .clip(0, None)   # ❗ chặn score >= 0
    .alias("implicit_score")
)
```

```python
df.select("implicit_score").describe()
```

```python
df.write_parquet(DATASET_DIR / "purchase_2024_ready_daxulyoutlier.parquet")
```

```python
pl.Config.set_tbl_rows(20)
pl.Config.set_tbl_cols(50)
pl.Config.set_fmt_str_lengths(120)

DATASET_DIR = Path("dataset")

purchase_path = DATASET_DIR / "purchase_2024_ready_daxulyoutlier.parquet"
item_path     = DATASET_DIR / "item_2024_ready.parquet"
user_path     = DATASET_DIR / "user_2024_user_ge2.parquet"

purchase_path.exists(), item_path.exists(), user_path.exists()
```

```python
lf_purchase = pl.scan_parquet(purchase_path)
lf_item     = pl.scan_parquet(item_path)
lf_user     = pl.scan_parquet(user_path)
```

```python
print("PURCHASE schema:\n", lf_purchase.schema)
print("\nITEM schema:\n", lf_item.schema)
print("\nUSER schema:\n", lf_user.schema)
```

```python
def show_shape(lf, name):
    rows = lf.select(pl.len()).collect().item()
    cols = len(lf.schema)
    print(f"{name}: {rows:,} rows × {cols} cols")

show_shape(lf_purchase, "PURCHASE")
show_shape(lf_item, "ITEM")
show_shape(lf_user, "USER")
```

```python
def null_report(lf, name, topk=20):
    rows = lf.select(pl.len()).collect().item()
    df_null = (
        lf.select([pl.col(c).null_count().alias(c) for c in lf.schema])
        .collect()
        .transpose(include_header=True, header_name="column", column_names=["nulls"])
        .with_columns((pl.col("nulls") / rows).alias("null_rate"))
        .sort("nulls", descending=True)
    )
    print(f"\n{name} null report (top {topk})")
    display(df_null.head(topk))

null_report(lf_purchase, "PURCHASE")
null_report(lf_item, "ITEM")
null_report(lf_user, "USER")
```

```python
USER_KEY = "customer_id"
ITEM_KEY = "item_id"

basic = lf_purchase.select([
    pl.len().alias("n_rows"),
    pl.col(USER_KEY).n_unique().alias("n_users"),
    pl.col(ITEM_KEY).n_unique().alias("n_items"),
]).collect()

basic
```

```python
# interactions per user / per item (describe)
per_user = (
    lf_purchase.group_by(USER_KEY)
    .agg(pl.len().alias("n_interactions"))
    .select("n_interactions")
    .collect()
    .describe()
)

per_item = (
    lf_purchase.group_by(ITEM_KEY)
    .agg(pl.len().alias("n_interactions"))
    .select("n_interactions")
    .collect()
    .describe()
)

per_user, per_item
```
