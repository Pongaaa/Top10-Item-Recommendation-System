```python
import pandas as pd
import json

# ====================================================
# 1. C·∫§U H√åNH (Kh√¥ng c·∫ßn Top Items n·ªØa nh∆∞ng c·ª© ƒë·ªÉ ƒë√¢y n·∫øu mu·ªën d√πng l·∫°i sau)
# ====================================================
# TOP_ITEMS_COLD_START = [...] # ƒê√£ comment l·∫°i v√¨ kh√¥ng d√πng ƒë·∫øn

# ====================================================
# 2. ƒê·ªåC D·ªÆ LI·ªÜU
# ====================================================
print("‚è≥ ƒêang ƒë·ªçc file Ground Truth...")
gt_df = pd.read_csv('final_groundtruth.csv')

# T·∫°o danh s√°ch c√°c User B·∫ÆT BU·ªòC (ƒë·ªÉ l·ªçc)
required_users = set(gt_df['customer_id'].astype(str))
print(f"‚úÖ T√¨m th·∫•y {len(required_users)} users c·∫ßn d·ª± ƒëo√°n trong file CSV.")

input_json = "pred_feb2025.json"
output_json = "submission (4).json" # ƒê·ªïi t√™n file output cho d·ªÖ ph√¢n bi·ªát

try:
    with open(input_json, 'r') as f:
        pred_data = json.load(f) 
    
    print(f"üìÇ ƒê√£ load file JSON g·ªëc. T·ªïng s·ªë user c√≥ d·ª± ƒëo√°n: {len(pred_data)}")
    
    # ====================================================
    # 3. X·ª¨ L√ù: L·ªåC & B·ªé QUA COLD START
    # ====================================================
    final_submission = {}
    skipped_count = 0 # Bi·∫øn ƒë·∫øm s·ªë l∆∞·ª£ng user b·ªã b·ªè qua
    
    # Duy·ªát qua t·ª´ng user trong danh s√°ch Ground Truth
    for user in required_users:
        if user in pred_data:
            # TR∆Ø·ªúNG H·ª¢P 1: C√≥ d·ª± ƒëo√°n -> L·∫•y k·∫øt qu·∫£
            final_submission[user] = pred_data[user]
        else:
            # TR∆Ø·ªúNG H·ª¢P 2: User thi·∫øu (Cold Start) -> B·ªé QUA (Kh√¥ng l√†m g√¨ c·∫£)
            skipped_count += 1
            
    # ====================================================
    # 4. L∆ØU K·∫æT QU·∫¢
    # ====================================================
    print(f"‚úÖ ƒê√£ x·ª≠ l√Ω xong!")
    print(f"   - T·ªïng user y√™u c·∫ßu (GT): {len(required_users)}")
    print(f"   - T·ªïng user c√≥ trong file k·∫øt qu·∫£: {len(final_submission)}")
    print(f"   - S·ªë user b·ªã B·ªé QUA (Cold-start): {skipped_count}")
    
    with open(output_json, 'w') as f:
        json.dump(final_submission, f)
        
    print(f"üíæ ƒê√£ l∆∞u file ho√†n ch·ªânh: {output_json}")
    
except FileNotFoundError:
    print(f"‚ùå Kh√¥ng t√¨m th·∫•y file '{input_json}'. H√£y ki·ªÉm tra l·∫°i t√™n file.")
```
