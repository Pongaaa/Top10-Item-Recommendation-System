```python
def coerce_numeric_columns(df, candidates=None):
    df = df.copy()
    if candidates is None:
        candidates = []
        for c in df.select_dtypes(include="object").columns:
            s = pd.to_numeric(df[c].str.replace(r"[^\d\.\-eE]", "", regex=True), errors="coerce")
            ok = s.notna().mean()
            if ok >= 0.9:
                candidates.append(c)
    for c in candidates:
        df[c] = pd.to_numeric(df[c].astype(str).str.replace(r"[^\d\.\-eE]", "", regex=True), errors="coerce")
    return df

def coerce_epoch_columns(df, cols, unit="s"):
    df = df.copy()
    for c in cols:
        if c in df.columns:
            df[c] = pd.to_datetime(df[c], unit=unit, errors="coerce")
    return df

def full_describe(df, percentiles=[.05,.25,.5,.75,.95]):
    out = {}

    num_cols = df.select_dtypes(include=[np.number]).columns
    if len(num_cols):
        desc_num = df[num_cols].describe(percentiles=percentiles).T
        desc_num["missing"] = df[num_cols].isna().sum()
        desc_num["missing_rate"] = (df[num_cols].isna().mean()*100).round(2)
        out["numeric"] = desc_num.sort_index()

    cat_cols = df.select_dtypes(include=["object","category","bool"]).columns
    if len(cat_cols):
        desc_cat = pd.DataFrame({
            "count": df[cat_cols].count(),
            "unique": df[cat_cols].nunique(dropna=True),
            "top": df[cat_cols].apply(lambda s: s.mode(dropna=True).iloc[0] if s.dropna().size else np.nan),
            "freq": df[cat_cols].apply(lambda s: s.value_counts(dropna=True).iloc[0] if s.dropna().size else np.nan),
            "missing": df[cat_cols].isna().sum(),
            "missing_rate": (df[cat_cols].isna().mean()*100).round(2)
        })
        out["categorical"] = desc_cat.sort_index()

    dt_cols = df.select_dtypes(include=["datetime64[ns]","datetime64[us]","datetime64[ms]"]).columns
    if len(dt_cols):
        desc_dt = pd.DataFrame({
            "min": df[dt_cols].min(),
            "max": df[dt_cols].max(),
            "missing": df[dt_cols].isna().sum(),
            "missing_rate": (df[dt_cols].isna().mean()*100).round(2)
        })
        out["datetime"] = desc_dt.sort_index()

    miss = pd.DataFrame({
        "dtype": df.dtypes.astype(str),
        "missing": df.isna().sum(),
        "missing_rate(%)": (df.isna().mean()*100).round(2),
        "nunique": df.nunique(dropna=True)
    }).sort_values("missing_rate(%)", ascending=False)
    out["missing_overview"] = miss

    return out

def show_sections(fdict, title_prefix):
    for k, v in fdict.items():
        print(f"\n==== {title_prefix} | {k.upper()} ====")
        display(v)
```

```python
def univariate_plots(df, num_cols, title_prefix="", max_bins=60, q_low=0.01, q_high=0.99):
    import numpy as np
    import matplotlib.pyplot as plt
    import seaborn as sns

    for col in num_cols:
        s = pd.to_numeric(df[col], errors="coerce").dropna()
        if s.size == 0 or s.nunique() <= 1:
            print(f"(B·ªè qua {col}: r·ªóng ho·∫∑c ch·ªâ 1 gi√° tr·ªã)")
            continue

        # Th·ªëng k√™ nhanh
        q1, q99 = s.quantile(q_low), s.quantile(q_high)
        skew = s.skew()
        ratio = (q99 - q1) / (s.max() - s.min() + 1e-9)

        # ƒêi·ªÅu ki·ªán d√πng log: gi√° tr·ªã >0, l·ªách ph·∫£i m·∫°nh
        use_log = (s.min() > 0) and (skew > 1.0) and (s.quantile(0.99) / max(s.median(), 1e-9) > 5)

        # 1) Histogram (zoom trung t√¢m b·∫±ng ph√¢n v·ªã)
        s_clip = s.clip(lower=q1, upper=q99)
        cut = (s < q1).sum() + (s > q99).sum()

        plt.figure(figsize=(7,4))
        sns.histplot(s_clip, bins="auto", kde=False)
        plt.title(f"{title_prefix} ‚Äì Histogram c·ªßa {col} (zoom {int(q_low*100)}‚Äì{int(q_high*100)}%)")
        plt.xlabel(col); plt.ylabel("T·∫ßn su·∫•t")
        if cut > 0:
            plt.suptitle(f"ƒê√£ c·∫Øt {cut:,} ƒëi·ªÉm ngo√†i kho·∫£ng [{q1:,.2f}, {q99:,.2f}]", y=0.98, fontsize=9)
        plt.tight_layout(); plt.show()

        # 2) M·∫≠t ƒë·ªô (KDE) tr√™n d·ªØ li·ªáu ƒë√£ zoom
        plt.figure(figsize=(7,4))
        sns.kdeplot(s_clip, fill=True)
        plt.title(f"{title_prefix} ‚Äì M·∫≠t ƒë·ªô (KDE) c·ªßa {col} (zoom {int(q_low*100)}‚Äì{int(q_high*100)}%)")
        plt.xlabel(col); plt.ylabel("M·∫≠t ƒë·ªô")
        plt.tight_layout(); plt.show()

        # 3) Box-plot (zoom)
        plt.figure(figsize=(7,2.8))
        sns.boxplot(x=s_clip, orient="h")
        plt.title(f"{title_prefix} ‚Äì Box-plot c·ªßa {col} (zoom)")
        plt.xlabel(col)
        plt.tight_layout(); plt.show()

        # 4) (Tu·ª≥ ch·ªçn) Phi√™n b·∫£n log n·∫øu l·ªách m·∫°nh
        if use_log:
            s_log = np.log1p(s)
            plt.figure(figsize=(7,4))
            sns.histplot(s_log, bins="auto", kde=False)
            plt.title(f"{title_prefix} ‚Äì Histogram (log1p) c·ªßa {col}")
            plt.xlabel(f"log1p({col})"); plt.ylabel("T·∫ßn su·∫•t")
            plt.tight_layout(); plt.show()

            plt.figure(figsize=(7,4))
            sns.kdeplot(s_log, fill=True)
            plt.title(f"{title_prefix} ‚Äì M·∫≠t ƒë·ªô (log1p) c·ªßa {col}")
            plt.xlabel(f"log1p({col})"); plt.ylabel("M·∫≠t ƒë·ªô")
            plt.tight_layout(); plt.show()
```
